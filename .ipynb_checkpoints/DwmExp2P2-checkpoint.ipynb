```python
# Data Manipulation libraries:
import io
import numpy as np
import pandas as pd
from google.colab import files
```

```python
df = files.upload()
```

```python
df = pd.read_csv("/content/olap_exp2.csv", encoding= 'unicode_escape')
```

```python
df
```

```python
df['year']=pd.to_datetime(df['OrderDate']).dt.year
df['month']=pd.to_datetime(df['OrderDate']).dt.month
df
```

#2.	Need to see monthly, quarterly, yearly profit, sales of each category region wise. 

```python
table = pd.pivot_table(df,values=['Sales','Profit'],index=['year','month','Category','Sub-Category','Region'],aggfunc = np.sum)
table
```

#3.	Comparison of sales and profit on various year. 

```python
cube=pd.pivot_table(df,values=['Sales','Profit'],index=['year'], aggfunc = np.sum)
cube
```

#4. Comparison of sales in various months for product name =x. 

```python
x = input("Enter the product name : ")
sales_comp = pd.pivot_table(df[df['Product Name'] == x], values='Sales', index=['month'], columns=['year'], aggfunc = np.sum)
sales_comp
```

#5.	Need to know which product has more demand on which location? 

```python
x = input("Enter the product name : ")
demand=pd.pivot_table(df[df['Product Name'] == x],values='Sales',index=['year','month','State','City'],columns=['Region'], aggfunc = np.sum)
demand
```

#6.	Need to study trend of sales by time period of the day over the month, and year?

```python
trends=pd.pivot_table(df,values='Sales',index=['year','month'], aggfunc = np.sum)
trends
```

#7.	What is sale of  product category wise, city wise for year=2015

```python
yr = int(input("Enter the year : "))
yearly=pd.pivot_table(df[df['year'] == yr],values='Sales',index=['month','City','Category','Product Name'], aggfunc = np.sum)
yearly
```

#8.	What is sales of product categories  of Jan  month of each year? 

```python
month = int(input("Enter the month number : "))
monthly=pd.pivot_table(df[df['month'] == month],values=['Sales'],index=['year','Category', 'Product Name'],aggfunc=np.sum)
monthly
```

#9.	What is trend of sales on year 2014 and 2015 for product category furniture and technology? 

```python
trend_year = pd.pivot_table(df, values=['Sales'], index=['year'], columns = ['Category'], aggfunc=np.sum)
trend_year
```

#10.	Need to compare weekly, monthly and yearly sales to know growth.

```python
growth = pd.pivot_table(df,values=['Sales'],index=['year','Category'], columns = ['month'],aggfunc = np.sum)
growth
```
