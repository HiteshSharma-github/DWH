#Q.6 Give visualization of statistical description of data â€“ in form of histogram, scatter plot, pie chart

##Histogram

```python
import matplotlib.pyplot as plt
import pandas as pd

fig=plt.figure()

ax = fig.add_subplot(1,1,1)

ax.hist(df['age'],bins = 5)

plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Patients')
plt.show()
```

##Scatter Plot

```python
fig=plt.figure()

ax = fig.add_subplot(1,1,1)
#Variable
ax.scatter(df.head(10)['age'],df.head(10)['cp'])
#Labels and Tit
plt.title('CP and Age distribution')
plt.xlabel('Age')
plt.ylabel('CP')
plt.show()
```

##Pie Chart

```python
df.groupby(['sex']).sum().plot(kind='pie', y='age' )
```
